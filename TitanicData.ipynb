```python
# Dataset 3 is connected to dataset 2 so merge these two data to perform analysis easily
## Both dataset have one common column 'PassengerId', this will allow you to combine them based on the common column, and you can control whether you want all rows (outer join), only matching rows (inner join), or all rows from one side (left or right join).
## on='PassengerId': Specifies the column to join on.
how='inner': Keeps only rows with matching values in common_column (inner join).
## The resulting combined_df will contain all columns from both datasets. Rows will be matched based on the common_column values.
```

```python
combined_df = pd.merge(df1, df2, on='PassengerId', how='inner') 
print(combined_df)
```

```python
# Import libraries to preprocess the data for analysis
## To process and clean columns like age and sex, and to detect and handle outliers in columns such as fare and age, follow these steps in Python with pandas.
##  Convert sex to numeric values: Map 'male' to 1 and 'female' to 0.
## Convert age to numeric format: Ensure that age is of numeric type and handle missing values.
```

```python
missing_values = comb_df.isnull().sum().sort_values(ascending=False)
missing_values_percentage = missing_values / len(comb_df) * 100
print("The percentage of missing values in dataset is:\n", missing_values_percentage.round(2))
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
```

```python


# impute missing values in age and fare
comb_df['Age'] = pd.to_numeric(comb_df['Age'], errors='coerce')
comb_df['Age'].fillna(comb_df['Age'].mean(), inplace=True)
comb_df['Fare'].fillna(comb_df['Fare'].median(), inplace=True)
# impute missing values in embark and cabin
comb_df['Cabin'] = comb_df['Cabin'].fillna(0)
comb_df['Embarked'] =comb_df['Embarked'].fillna(0)
comb_df['Sex'] = comb_df['Sex'].map({'male': 1, 'female': 0})

# encode the categorical variables using for loop where object and categoy datatypes are given
for col in comb_df.columns:
    if comb_df[col].dtype == 'object' or comb_df[col].dtype.name == 'category':
        comb_df[col] = LabelEncoder().fit_transform(comb_df[col])
comb_df.head()
```

```python
# We can calculate survival rates grouped by Pclass and Sex.
# Create a bar plot to visualize the survival rates based on passenger class and Sex.
```

```python
survival_rates = comb_df.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()
survival_rates.columns = ['Pclass', 'Sex', 'Survival Rate']
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=survival_rates, x='Pclass', y='Survival Rate', hue='Sex')
plt.title('Survival Rate by Pclass and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.ylim(0, 1)
plt.legend(title='Sex')
plt.show()
```

```python
# To find correlations between numerical variables and the target variable (e.g., 'Survived')
```

```python
# Correlation matrix
correlation_matrix = comb_df.corr()

# Focus on correlation with 'Survived'
survived_correlation = correlation_matrix['Survived'].sort_values(ascending=False)
print(survived_correlation)

```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Summary and Key Insights:

## This correlation matrix visualizes the relationships between different variables in a dataset. The values in each cell range from -1 to 1, representing the strength and direction of the correlation between two variables:

## Positive Correlation (values closer to 1): A high value in one variable is associated with a high value in another. For instance, "Parch" and "SibSp" have a moderately strong positive correlation (0.37), meaning that having more siblings/spouses (SibSp) is somewhat associated with having more parents/children (Parch).
## Negative Correlation (values closer to -1): A high value in one variable is associated with a low value in another. For example, "Survived" and "Sex" have a strong negative correlation (-0.69), suggesting that gender (perhaps being male, in this case) is associated with lower survival rates.
## Weak Correlation (values closer to 0): There is little to no linear relationship between the two variables. For instance, "Fare" and "Age" have a weak positive correlation (0.18), meaning there's almost no relationship between them.
# Key Insights from this Matrix
## Survival and Gender: The correlation between "Survived" and "Sex" is -0.69, indicating a significant relationship where one gender is more likely to survive than the other (likely females).

## Survival and Class (Pclass): "Survived" and "Pclass" have a correlation of -0.26, suggesting a moderate negative relationship, where higher-class passengers (lower "Pclass" values) tend to have higher survival rates.

## Survival and Fare: Theres a moderate positive correlation (0.23) between "Fare" and "Survived," indicating that passengers who paid higher fares had better chances of survival.

## Class and Fare: Theres a strong negative correlation (-0.56) between "Pclass" and "Fare," which makes sense since higher-class tickets (lower "Pclass" values) would generally cost more.






```
